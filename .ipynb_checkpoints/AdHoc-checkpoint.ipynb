#  Lola Tamis: Customer Loyalty Analytics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
```

```python
# Read and display JSON file using Pandas

filename = "transaction-data-adhoc-analysis.json"
df = pd.read_json(filename)
df
```

```python
new_df = df.copy().dropna()
new_df.drop(columns=["username", "address", "birthdate", "mail", "sex"], inplace=True)

new_df
```

## Average Order Value (AOV)

The **Average Order Value** is obtained by dividing the gross revenue (total income sans expenses, debts, additional income streams, and operating costs) by the total number of orders.

```python
# Monthly basis
new_df.transaction_date = pd.to_datetime(new_df.transaction_date)
monthly_AOV = pd.DataFrame(new_df.resample('m', on='transaction_date')["transaction_value"].mean())
monthly_AOV
```

```python
# Yearly basis (half-year in the given dataset)
new_df.resample('y', on='transaction_date')["transaction_value"].mean()
yearly_AOV = pd.DataFrame(new_df.resample('y', on='transaction_date')["transaction_value"].mean())
yearly_AOV
```

## Monthly Total Item Sales and Frequency

First obtain the cost of each individual item. Here we see that **Candy City,Gummy Worms** is the least expensive item while **HealthyKid 3+,Nutrional Milk** is the most expensive item purchased by customers in Lola Tamis.

```python
cost_df = df.copy()

# Take only the relevant columns: the items to be extracted and their corresponding costs
cost_df = cost_df[["transaction_items", "transaction_value"]]

# Filter out the lists of items which do not exactly contain a single item.
cost_df["transaction_items"] = cost_df["transaction_items"].str.split(";")
cost_df = cost_df[cost_df["transaction_items"].apply(len) == 1]

# Get the quantity of each individual item and place it into a separate column. 
cost_df["transaction_items"] = cost_df["transaction_items"].apply(lambda x: x[0][:-1].split(",(x"))
cost_df["quantity"] = cost_df["transaction_items"].apply(lambda x: int(x[1]))

# Only retain the name of each item in the transaction_items column
cost_df["transaction_items"] = cost_df["transaction_items"].apply(lambda x: x[0].strip())

# Drop duplicate items
cost_df.drop_duplicates(subset="transaction_items", inplace=True)

# Mutate the transaction_value column by dividing each value (cost) by the corresponding quantity of purchased items.
cost_df["transaction_value"] /= cost_df["quantity"]

cost_df.set_index(["transaction_items"], inplace=True)
cost_df.drop(columns=["quantity"], inplace=True)
cost_df = cost_df.squeeze()
cost_df.sort_values()
```

```python
item_df = new_df.copy()

# Break down the transaction_items into lists of individual items
item_df["transaction_items"] = item_df["transaction_items"].str.split(";")

# Transform each element of the lists to a row, replicating index values (Python Data Analysis Library)
item_df = item_df.explode("transaction_items").reset_index(drop=True)

# Separate the item from the integer quantity
item_df["transaction_items"] = item_df["transaction_items"].apply(lambda x: x[:-1].split(',(x'))

# Assign item and quantity columns accordingly
item_df["quantity"] = item_df["transaction_items"].apply(lambda x: int(x[1]))
item_df["transaction_items"] = item_df["transaction_items"].apply(lambda x: x[0])

# Re-evaluate transaction values based on individual costs and quantity
item_df["transaction_value"] = item_df["quantity"] * item_df["transaction_items"].apply(lambda x: cost_df.loc[x])
item_df
```

```python
monthly_sales = pd.DataFrame(item_df[["transaction_value", "quantity"]].groupby([item_df.transaction_items, item_df.transaction_date.dt.month]).sum())
monthly_sales.index.names = ["Item", "Month"]
monthly_sales.columns = ["Sales", "Count"]
monthly_sales
```

## Customer Activity Status

```python
retention_df = item_df[['name','quantity']]
retention_df["month"] = item_df.transaction_date.dt.month
table = pd.pivot_table(retention_df, values='quantity', index='month', 
                       columns ='name', aggfunc=sum, fill_value=0)
table
```

```python
# In applying lambda functions, we sum over Booleans since they also evaluate to 0 or 1

repeater = table.apply(lambda x: [i != 1 and all([x[i-1],x[i]]) for i in x.index]).sum(axis=1)
inactive = table.apply(lambda x: [i != 1 and any(x[:i]) and not x[i] for i in x.index]).sum(axis=1)
engaged = table.apply(lambda x: [all(x[:i]) for i in x.index]).sum(axis=1)


activity_status = pd.DataFrame({'repeater': repeater, 'inactive': inactive, 'engaged': engaged})
activity_status
```

```python
# Plot activity status of customers over the course of six months as a multiple bar graph

ax = activity_status.plot.bar(rot=0)
ax
```
